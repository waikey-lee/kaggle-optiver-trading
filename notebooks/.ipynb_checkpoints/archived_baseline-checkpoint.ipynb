## Clipping

Can update this clipping logic
- To by group clipping instead of global clipping, because some stock might be always nearer to the clipping bound, while some are not
- By defining the bound using outlier factor instead of simple percentile

```python
PRICE_CLIP_PERCENTILE = 0.01 # 1 - (2 * PRICE_CLIPPER_TAIL / 100) will remained unclip
```

```python
def get_price_clippers(df, price_cols, price_clip_percentile=PRICE_CLIP_PERCENTILE):
    price_clippers = {}
    for price_col in price_cols:
        upper_bound = np.percentile(df[price_col].dropna(), 100 - price_clip_percentile)
        lower_bound = np.percentile(df[price_col].dropna(), price_clip_percentile)
        price_clippers[price_col] = (lower_bound, upper_bound)
        cprint(f"For {price_col}, the global clip bound is", end=" ", color="blue")
        cprint(f"({lower_bound:.4f}, {upper_bound:.4f})", color="green")
    return price_clippers
```

```python
base_price_cols = get_cols(train, contains="price")
price_clippers = get_price_clippers(train, base_price_cols)
```

```python
VOLUME_CLIPPER_UPPER_TAIL = 0.01 # only the top VOLUME_CLIPPER_UPPER_TAIL proportion will be clipped
```

```python
def get_volume_clippers(df, volume_cols, volume_clip_upper_percentile=VOLUME_CLIPPER_UPPER_TAIL):
    volume_clippers = {}
    for volume_col in volume_cols:
        upper_bound = np.percentile(df[volume_col].dropna(), 100 - volume_clip_upper_percentile)
        volume_clippers[volume_col] = (0, upper_bound)
        cprint(f"For {volume_col}, the global clip bound is", end=" ", color="blue")
        cprint(f"(0, {upper_bound:.4f})", color="green")
    return volume_clippers
```

```python
base_volume_cols = get_cols(train, contains="size")
volume_clippers = get_volume_clippers(train, base_volume_cols)
```

```python
PRICE_MULTIPLIER = 10 # Factor to multiple the price col after taking the distance to 1
PRICE_POWER_FACTOR = 1/2 # Factor to transform the price col to bimodal distribution, the lower the further apart of the modes
```

```python
MIN_TARGET = -100
MAX_TARGET = 100
```

```python
def clip(df, price_clippers=None, volume_clippers=None):
    # Find the column list for each group
    price_cols = get_cols(df, endswith="price")
    volume_cols = get_cols(df, endswith="size")
    flag_cols = get_cols(df, endswith="flag")
    base_cols = price_cols + volume_cols + flag_cols
    
    if price_clippers is None:
        price_clippers = get_price_clippers(df, price_cols)
    
    if volume_clippers is None:
        volume_clippers = get_volume_clippers(df, volume_cols)
    
    # Clip price columns
    for price_col in price_cols:
        df[price_col] = df[price_col].clip(*price_clippers[price_col])
        
    # Clip volume columns
    for volume_col in volume_cols:
        df[volume_col] = df[volume_col].clip(*volume_clippers[volume_col])
    
    # Clip target columns (if applicable)
    if "target" in df.columns:
        df["clipped_target"] = df["target"].clip(MIN_TARGET, MAX_TARGET)
        df["binary_target"] = (df["target"] > 0).astype(int)
    return df
```

```python
def get_lag_features(df):
    for col in columns_to_shift:
        if col.endswith("_price"):
            df[f"{col}_l1_ratio"] = df[col] / df[f"{col}_l1"]
            df[f"{col}_l2_ratio"] = df[col] / df[f"{col}_l2"]
            df[f"{col}_l1_l2_ratio_diff"] = df[f"{col}_l1_ratio"] - df[f"{col}_l2_ratio"]

            ori = (df[col] - 1)
            l1 = (df[f"{col}_l1"] - 1)
            l2 = (df[f"{col}_l2"] - 1)
            df[f"{col}_sc_l1_ratio"] = np.exp(ori) / np.exp(l1)
            df[f"{col}_sc_l2_ratio"] = np.exp(ori) / np.exp(l2)
            df[f"{col}_sc_l1_l2_ratio_diff"] = df[f"{col}_sc_l1_ratio"] - df[f"{col}_sc_l2_ratio"]
            
        elif col.endswith("_size") and not col.startswith("log_"):
            df[f"{col}_l1_ratio"] = df[col] / (df[col] + df[f"{col}_l1"])
            df[f"{col}_l2_ratio"] = df[col] / (df[col] + df[f"{col}_l2"])
            df[f"{col}_l1_l2_ratio_diff"] = (df[f"{col}_l1_ratio"] + df[f"{col}_l2_ratio"]) / 2
            
    return df
```

```python
def compute_lag_features(df):
    for col in tqdm(columns_to_shift):
        if col.endswith("_price") and col.startswith("ori_"):
            df[f"{col}@lag1_ratio"] = df[col] / df[f"{col}@lag1"]
            df[f"{col}@lag2_ratio"] = df[col] / df[f"{col}@lag2"]
            df[f"{col}@lag12_ratio_diff"] = df[f"{col}@lag1_ratio"] - df[f"{col}@lag2_ratio"]
        elif col.endswith("_price") and col.startswith("st-"):
            df[f"{col}@lag1_ratio"] = np.exp(df[col]) / np.exp(df[f"{col}@lag1"])
            df[f"{col}@lag2_ratio"] = np.exp(df[col]) / np.exp(df[f"{col}@lag1"])
            df[f"{col}@lag12_ratio_diff"] = df[f"{col}@lag1_ratio"] - df[f"{col}@lag2_ratio"]
        elif col.endswith("_size"):
            df[f"{col}@lag1_ratio"] = df[col] / (1 + df[f"{col}@lag1"])
            df[f"{col}@lag2_ratio"] = df[col] / (1 + df[f"{col}@lag2"])
            df[f"{col}@lag12_ratio_diff"] = df[f"{col}@lag1_ratio"] - df[f"{col}@lag2_ratio"]
        elif col.endswith("flag"):
            df[f"{col}@lag1_diff"] = (df[col] - df[f"{col}@lag1"])
            df[f"{col}@lag2_diff"] = (df[col] - df[f"{col}@lag2"])
    return df
```

```python
def clip_and_transform(df):
    df = df.sort_values(by=["stock_id", "date_id", "seconds"]).reset_index(drop=True)
    
    price_cols = get_cols(df, endswith="price")
    volume_cols = get_cols(df, endswith="size")
    flag_cols = get_cols(df, endswith="flag")
    base_cols = price_cols + volume_cols + flag_cols
    
    # Clip & Transform price columns
    for price_col in price_cols:
        # Apply sqrt transformation on 6 price columns
        df[price_col] = (df[price_col].clip(*price_clippers[price_col]) - 1) * PRICE_MULTIPLIER
        df["sqrt_" + price_col] = my_power((df[price_col] - 1) * PRICE_MULTIPLIER, PRICE_POWER_FACTOR)
        
    # Clip & Transform volume columns
    for volume_col in volume_cols:
        # Apply log transformation on 4 volume columns
        df[volume_col] = df[volume_col].clip(*volume_clippers[volume_col])
        df["log_" + volume_col] = df[volume_col].apply(np.log1p)
    
    df.rename(columns={col: f"ori_{col}" for col in base_cols}, inplace=True)
    
    # Clip target columns (if applicable)
    if "target" in df.columns:
        df["clipped_target"] = df["target"].clip(MIN_TARGET, MAX_TARGET)
        df["binary_target"] = (df["target"] > 0).astype(int)
    return df
```

```python
def volume_interaction(df):
    suffix = "size"
    base_volume_columns = get_cols(df, contains="size", excludes=["log", "st"])
    log_volume_columns = list_diff(get_cols(df, contains=["log", "size"], excludes=["st"]), base_volume_columns)
    scaled_volume_columns = get_cols(df, startswith="st", endswith="size", excludes="log")
    scaled_log_volume_columns = get_cols(df, startswith="st", endswith="size", contains="log")
    # cprint(base_volume_columns, color="blue")
    # cprint(log_volume_columns, color="green")
    # cprint(scaled_volume_columns, color="blue")
    # cprint(scaled_log_volume_columns, color="green")
    
    for c1, c2 in combinations(base_volume_columns, 2):
        # c1 = col1.replace("_" + suffix, "")
        # c2 = col2.replace("_" + suffix, "")
        df[f'{c1}&{c2}#sum'] = df[c1] + df[c2]
        df[f'{c1}&{c2}#diff'] = df[c1] - df[c2]
        df[f'{c1}&{c2}#ratio'] = df[c1] / df[c2]
        df[f'{c1}&{c2}#ds_ratio'] =df[f'{c1}&{c2}#diff'] / df[f'{c1}&{c2}#sum']
        
    for c1, c2 in combinations(log_volume_columns, 2):
        # c1 = col1.replace("_" + suffix, "")
        # c2 = col2.replace("_" + suffix, "")
        df[f'{c1}&{c2}#ratio'] = df[c1] / df[c2]
        
    for col1, col2 in combinations(scaled_volume_columns, 2):
        # c1 = col1.replace("_" + suffix, "")
        # c2 = col2.replace("_" + suffix, "")
        df[f'{c1}&{c2}#sum'] = df[c1] + df[c2]
        df[f'{c1}&{c2}#diff'] = df[c1] - df[c2]
        df[f'{c1}&{c2}#ratio'] = df[c1] / df[c2]
        df[f'{c1}&{c2}#ds_ratio'] =df[f'{c1}&{c2}#diff'] / df[f'{c1}&{c2}#sum']

    for col1, col2 in combinations(scaled_log_volume_columns, 2):
        # c1 = col1.replace("_" + suffix, "")
        # c2 = col2.replace("_" + suffix, "")
        df[f'{c1}&{c2}#sum'] = df[c1] + df[c2]
        df[f'{c1}&{c2}#diff'] = df[c1] - df[c2]
        df[f'{c1}&{c2}#ratio'] = df[c1] / df[c2]
        df[f'{c1}&{c2}#ds_ratio'] =df[f'{c1}&{c2}#diff'] / df[f'{c1}&{c2}#sum']
        
    # df.drop(columns=noob_volume_interaction_features, inplace=True)
    return df
```

```python
def price_interaction(df):
    suffix = "price"
    base_price_columns = get_cols(df, contains="price", excludes=["sqrt", "st"])
    sqrt_price_columns = list_diff(get_cols(df, contains=["sqrt", "price"], excludes=["st"]), base_price_columns)
    scaled_price_columns = get_cols(df, startswith="st", endswith="price", excludes="sqrt")
    scaled_sqrt_price_columns = get_cols(df, startswith="st", endswith="price", contains="sqrt")
    # cprint(base_price_columns, color="blue")
    # cprint(sqrt_price_columns, color="green")
    # cprint(scaled_price_columns, color="blue")
    # cprint(scaled_sqrt_price_columns, color="green")
    
    for c1, c2 in combinations(base_price_columns, 2):
        df[f'{c1}&{c2}#avg'] = (df[c1] + df[c2]) / 2
        df[f'{c1}&{c2}#diff'] = df[c1] - df[c2]
        df[f'{c1}&{c2}#ratio'] = df[c1] / df[c2]
        df[f'{c1}&{c2}#da_ratio'] =df[f'{c1}&{c2}#diff'] / df[f'{c1}&{c2}#avg']
        
    for c1, c2 in combinations(sqrt_price_columns, 2):
        df[f'{c1}&{c2}#avg'] = (df[c1] + df[c2]) / 2
        df[f'{c1}&{c2}#diff'] = df[c1] - df[c2]
        df[f'{c1}&{c2}#ratio'] = df[c1] / df[c2]
        df[f'{c1}&{c2}#da_ratio'] =df[f'{c1}&{c2}#diff'] / df[f'{c1}&{c2}#avg']
        
    for c1, c2 in combinations(scaled_price_columns, 2):
        df[f'{c1}&{c2}#avg'] = (df[c1] + df[c2]) / 2
        df[f'{c1}&{c2}#diff'] = df[c1] - df[c2]
        df[f'{c1}&{c2}#ratio'] = df[c1] / df[c2]
        df[f'{c1}&{c2}#da_ratio'] =df[f'{c1}&{c2}#diff'] / df[f'{c1}&{c2}#avg']
        
    for c1, c2 in combinations(scaled_sqrt_price_columns, 2):
        df[f'{c1}&{c2}#avg'] = (df[c1] + df[c2]) / 2
        df[f'{c1}&{c2}#diff'] = df[c1] - df[c2]
        df[f'{c1}&{c2}#ratio'] = df[c1] / df[c2]
        df[f'{c1}&{c2}#da_ratio'] =df[f'{c1}&{c2}#diff'] / df[f'{c1}&{c2}#avg']
        
    return df
```

```python
def price_volume_interaction(df):
    selected_price_cols = get_cols(df, contains=["bid", "ask"], endswith="price", startswith="ori")
    selected_size_cols = get_cols(df, contains=["bid", "ask"], endswith="size")
    for p, s in product(selected_price_cols, selected_size_cols):
        df[f'{p}&{s}#product'] = df[p] * df[s]
    
    pv_product_cols = df.columns[df.columns.str.contains(r"price\&.+size\#product", regex=True)].tolist()
    for pv1, pv2 in tqdm(combinations(pv_product_cols, 2)):
        # Remove the #agg tag, then split by & tag, format => [price]&[volume]#[agg]
        price_col1, volume_col1 = pv1.split("#")[0].split("&")
        price_col2, volume_col2 = pv2.split("#")[0].split("&")

        # price or volume feature has the format of [scaled type]-[transform type]-[base]
        same_price_process = price_col1.split("-")[:1] == price_col2.split("-")[:1]
        diff_price_base = price_col1.split("-")[-1] != price_col2.split("-")[-1]

        same_volume_process = volume_col1.split("-")[:1] == volume_col2.split("-")[:1]
        diff_volume_base = volume_col1.split("-")[-1] != volume_col2.split("-")[-1]

        # cprint(f"{same_price_process}, {diff_price_base}, {same_volume_process}, {diff_volume_base}", color="blue")
        if all([same_price_process, diff_price_base, same_volume_process, diff_volume_base]):
            df[f'{pv1}&&{pv2}##sum'] = (df[pv1] + df[pv2])
            df[f'{pv1}&&{pv2}##diff'] = df[pv1] - df[pv2]
            df[f'{pv1}&&{pv2}##da_ratio'] =df[f'{pv1}&&{pv2}##diff'] / df[f'{pv1}&&{pv2}##sum']
    return df
```

```python
def feature_engineering(df):
    df = df.merge(median_vol, how="left", left_on="stock_id", right_index=True)
    
    size_cols = get_cols(df, endswith="size", contains="ori")
    ori_price_cols = get_cols(df, endswith="price", contains="ori")
    normal_price_cols = get_cols(df, endswith="price", contains="ori", excludes=["sqrt", "far", "near"])
    sc_price_cols = get_cols(df, endswith="price", contains=["sqrt"])
    
    for s1, s2 in combinations(size_cols, 2):
        df[f'{s1}_{s2}_sum'] = df[s1] + df[s2]
        df[f'{s1}_{s2}_diff'] = df[s1] - df[s2]
        df[f'{s1}_{s2}_ratio'] = df[s1] / df[s2]
        df[f'{s1}_{s2}_ratio_sign'] = df[s1] / df[s2] * df['ori_imb_flag']
        df[f'{s1}_{s2}_log_ratio'] = np.log(df[s1]) / np.log(df[s2])
        df[f'{s1}_{s2}_log_ratio_sign'] = np.log(df[s1]) / np.log(df[s2]) * df['ori_imb_flag']
        df[f'{s1}_{s2}_diff_sum_ratio'] = df[f'{s1}_{s2}_diff'] / df[f'{s1}_{s2}_sum']
        df[f'{s1}_{s2}_diff_sum_ratio_sign'] = df[f'{s1}_{s2}_diff'] / df[f'{s1}_{s2}_sum'] * df['ori_imb_flag']
    
    for s, p in product(size_cols, normal_price_cols):
        df[f'{s}_{p}_sp_product'] = df[s] * df[p]
        
    sp_product_columns = get_cols(df, contains="_sp_product")
    for sp1, sp2 in combinations(sp_product_columns, 2):
        t1 = sp1.replace("_sp_product", "")
        t2 = sp2.replace("_sp_product", "")
        df[f'{t1}_{t2}_sp_product_diff'] = df[sp1] - df[sp2]
        
    df['mean_size'] = df['stock_id'].map(median_sizes.to_dict())
    df['std_size'] = df['stock_id'].map(std_sizes.to_dict()) 
    
    for price_cols in [ori_price_cols, sc_price_cols]:
        for p1, p2 in combinations(price_cols, 2):

            df[f'{p1}_{p2}_diff'] = (df[p1] - df[p2]).astype(np.float32)
            df[f'{p1}_{p2}_sum'] = (df[p1] + df[p2]).astype(np.float32)
            df[f'{p1}_{p2}_product'] = (df[p1] * df[p2]).astype(np.float32)
            df[f'{p1}_{p2}_ratio'] = (np.exp(df[p1]) / np.exp(df[p2])).astype(np.float32)
            df[f'{p1}_{p2}_imb'] = df[f'{p1}_{p2}_diff'] / df[f'{p1}_{p2}_sum']

    for c in combinations(ori_price_cols, 3):

        max_ = df[list(c)].max(axis=1)
        min_ = df[list(c)].min(axis=1)
        mid_ = df[list(c)].sum(axis=1) - min_ - max_

        df[f'{c[0]}_{c[1]}_{c[2]}_imb2'] = (max_ - mid_) / (mid_ - min_)
    
    df = pd.concat(
        [df] + [df.groupby(["stock_id", "date_id"])[columns_to_shift].shift(d).add_suffix(f"_lag{d}") for d in range(1, 3)], 
        axis=1
    )
    
    for col in columns_to_shift:
        if col.endswith("_price") and not col.startswith("ori-sqrt"):
            df[f"{col}_lag1_ratio"] = df[col] / df[f"{col}_lag1"]
            df[f"{col}_lag2_ratio"] = df[col] / df[f"{col}_lag2"]
            df[f"{col}_lag1_lag2_ratio_diff"] = df[f"{col}_lag1_ratio"] - df[f"{col}_lag2_ratio"]
        elif col.endswith("_price") and col.startswith("ori-sqrt"):
            df[f"{col}_lag1_ratio"] = np.exp(df[col]) / np.exp(df[f"{col}_lag1"])
            df[f"{col}_lag2_ratio"] = np.exp(df[col]) / np.exp(df[f"{col}_lag2"])
            df[f"{col}_lag1_lag2_ratio_diff"] = df[f"{col}_lag1_ratio"] - df[f"{col}_lag2_ratio"]
        elif col.endswith("_size"):
            df[f"{col}_lag1_ratio"] = np.log1p(df[col] / df[f"{col}_lag1"])
            df[f"{col}_lag2_ratio"] = np.log1p(df[col] / df[f"{col}_lag2"])
            df[f"{col}_lag1_lag2_ratio_diff"] = df[f"{col}_lag1_ratio"] - df[f"{col}_lag2_ratio"]
        elif col.endswith("flag"):
            df[f"is_increase"] = (df[col] > df[f"{col}_lag1"]).astype(int)
        
    gc.collect()
    
    return df
```

```python
# Inference on whole dataset, by batch
def inference_full(model, data, batch_size=8192):
    score_list = []
    for i in tqdm(range(int(data.shape[0] / batch_size) + 1)):
        try:
            score_list.extend(model.predict(
                data[model.feature_name()].iloc[int(i * batch_size): int((i+1) * batch_size)], 
            ))
        except:
            print("Too Many Batch")
    return score_list
```

```python
def calc_score(y_pred, df, is_zero_sum=True):
    pred_df = pd.DataFrame(dict(gt=df["target"], pred=y_pred, bid_size=df["bid_size"], ask_size=df["ask_size"]))
    pred_df.dropna(subset=["gt"], inplace=True)
    if is_zero_sum:
        score = (zero_sum(pred_df["pred"], pred_df["bid_size"] + pred_df["ask_size"]) - pred_df["gt"]).abs().mean()
    else:
        score = (pred_df["pred"] - pred_df["gt"]).abs().mean()
    return score
```
